```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bsoup
import json
import requests
import warnings

warnings.filterwarnings('ignore')
pd.options.display.max_columns = 50
```

```python
# set the base url for the request
team_abbr = 'STL'
season = '2021'
url = f'https://www.baseball-reference.com/teams/{team_abbr}/{season}-schedule-scores.shtml'

response = requests.get(url)


# while pages < limit:
#   clear_output(wait=True)

if (response.ok):
    data = response.content
    soup = bsoup(data, 'html.parser')

  # code for extracting the number of pages in the search and setting it equal to the max limit
  # find the div containing the page number our of the total number of pages
#   page_count_raw = 
    tag_list = soup.select('a[title="Click for standings on this date"]')
    game_dates = []
    game_links = []
    for tag in tag_list:
        game_dates.append(str(tag)[22:32])
    boxscore_links = soup.find_all('a', string='boxscore')
    link_url = 'https://www.baseball-reference.com'
    for link in boxscore_links:
        game_links.append((link_url+str(link)[9:38]))
game_dates
game_links

regular_season = dict(list(zip(game_dates, game_links)))
regular_season
```
